# Libraries

```python
import pandas as pd
import numpy as np
```

# Data

```python
# parse the log file and create dataframe

df = pd.DataFrame(columns = ["patient_id", "curative_palliative_status"])

theFile = open("../data/treat_type_tfk_1.15.2022.log","r")
FILE = theFile.readlines()
theFile.close()
printList = []
for i, line in enumerate(FILE[11:]):
    try:
        if len(line.split("|")[1].strip().split("   ")) < 2: # breaks in table
            continue
        elif len(line.split("|")[1].strip().split("   ")) == 2:
            pid, status = line.split("|")[1].strip().split("   ")
            pid = float(pid)
        else: # missing data
            pass
    except:
        pass
    df.loc[i, ["patient_id", "curative_palliative_status"]] = pid, status
```

```python
df = df.drop_duplicates()
print(df.shape)
df["patient_id"].nunique()
```

```python
data = pd.read_csv("../../../data/data.csv")

large_data = data.loc[:861, :]
large_data["patient_id"] = large_data["patient_id"].astype(float)
small_data = data.loc[862:, :]

large_data.head()
```

```python
large_data = large_data.merge(df, how="left", on="patient_id")
```

```python
print(large_data.shape)
print(large_data[~large_data["curative_palliative_status"].isnull()].shape)

small_data["curative_palliative_status"] = np.nan
```

```python
data = pd.concat([large_data, small_data], axis=0)
print(data.shape)
```

```python
data.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
data.to_csv("../../../data/data.csv", index=False)
```
